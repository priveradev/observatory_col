```python
# lote en barichara
# https://www.puntopropiedad.com/inmueble/ba4a-b71e-1938b1c-194a5d424d76-7e55
precio_lote_barichara = 170_000_000
# relacion de avaluo catastral vs precio comercial (Gemini)
relacion_avaluo_valor = 0.6
avaluo_lote_barichara = precio_lote_barichara * relacion_avaluo_valor
# incremento en barichara del avaluo catastral 4877%
# https://www.igac.gov.co/sites/default/files/documentos/351.%202057-2025%20Art%2049.pdf
incremento_avaluo_catastro = 11717
nuevo_avaluo_lote_barichara = avaluo_lote_barichara * (1 + (incremento_avaluo_catastro/100))
print(f'viejo avaluo: {avaluo_lote_barichara:_.0f}, nuevo avaluo: {nuevo_avaluo_lote_barichara:_.0f}')
```

```python
# tope de declaración de renta 224.095.500 (Gemini)
tope_patrimonio_declaracion_renta = 224_095_500 
pago_renta_antes = 'no pagas renta' if avaluo_lote_barichara < tope_patrimonio_declaracion_renta else 'pagas renta'
pago_renta_ahora = 'no pagas renta' if nuevo_avaluo_lote_barichara < tope_patrimonio_declaracion_renta else 'pagas renta'
# tope de impuesto al patrimonio 3.882.060.000 (Gemini)
tope_patrimonio_impuesto_patrimonio = 3_882_060_000
pago_patrimonio_antes = 'no pagas renta' if avaluo_lote_barichara < tope_patrimonio_impuesto_patrimonio else 'pagas renta'
pago_patrimonio_ahora = 'no pagas impuesto de patrimonio' if nuevo_avaluo_lote_barichara < tope_patrimonio_impuesto_patrimonio else 'pagas impuesto patrimonio'
print('IMPUESO DE RENTA')
print(f'antes {pago_renta_antes}, ahora {pago_renta_ahora}')
print(f'el valor del lote es {nuevo_avaluo_lote_barichara/tope_patrimonio_declaracion_renta} veces el tope para declarar renta')
print('IMPUESTO DE PATRIMONIO')
print(f'antes {pago_patrimonio_antes}, ahora {pago_patrimonio_ahora}')
print(f'el lote representa el {nuevo_avaluo_lote_barichara/tope_patrimonio_impuesto_patrimonio} veces el tope para pagar impuesto al patrimonio')
```
